```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction.xrd import XRDCalculator

with MPRester("wWWfKUMvMZEcIFh8yWJb") as m:
    structure=m.get_structure_by_material_id("mp-866056")
    w = CifWriter(structure)
    w.write_file('AcN_mp-866056_symmetrized.cif')
```

```python
xrdcal = XRDCalculator()
xrd = xrdcal.get_pattern(structure)
print(xrd)
print('----------------------------------------------------------------------------------------')
print('les 3 premiers pics du diffractogramme vallent: 30.20504937, 30.20709784 et 30.20486465 ')
print('----------------------------------------------------------------------------------------')
print('indices de Miller associés:')
print(xrd.hkls[0])
print(xrd.hkls[1])
print(xrd.hkls[2])
```
